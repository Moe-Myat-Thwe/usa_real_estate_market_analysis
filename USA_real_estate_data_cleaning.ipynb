```python
# Import the libraries
import pandas as pd
import numpy as np
```

```python
# Load the data
df = pd.read_csv(r"C:\Users\ASUS\OneDrive\Data Analytic Projects\USA Real Estate Dataset.csv", low_memory=False)
```

```python
df.shape
```

```python
# Look at the data
df.head(5)
```

```python
df.info()
df.describe(include='all')
```

```python
# Standardize column names
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_')
      .str.replace('-', '_'))
```

```python
# Drop useless unnamed columns
df.drop(['unnamed:_12', 'unnamed:_13', 'unnamed:_14'], axis=1, inplace=True)
```

```python
#check duplicates
df.duplicated().sum()
```

```python
df.dtypes
```

```python
# Strip & standardize strings
object_cols = df.select_dtypes(include="object").columns
df[object_cols] = df[object_cols].apply(lambda x: x.astype("string").str.strip())
```

```python
# Convert numeric columns properly
numeric_cols = ["price", "bed", "bath", "acre_lot", "house_size"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col])
```

```python
int_cols = ["price", "bed", "bath", "house_size"]

for col in int_cols:
    df[col] = df[col].round().astype("Int64")
```

```python
# Handle ID-like columns
df["brokered_by"] = df["brokered_by"].astype("Int64")
df["street"] = df["street"].astype("Int64")
df["zip_code"] = df["zip_code"].astype("Int64")
```

```python
df["prev_sold_date"] = pd.to_datetime(
    df["prev_sold_date"]
)
```

```python
#check missing values
df.info()
```

```python
#Identify critical columns
critical_cols = ["price", "state", "status"]

#Drop rows with nulls only in these columns
df = df.dropna(subset=critical_cols)
```

```python
#Columns to impute (replace nulls)
df["bed"] = df["bed"].fillna(df["bed"].median())
df["bath"] = df["bath"].fillna(df["bath"].median())
df["city"] = df["city"].fillna("Unknown")
# Acre & house size: keep NaN (important signal)
```

```python
int_type_cols = df.select_dtypes(include="int").columns
print(int_type_cols)

# Check for impossible values
print("column : shape")
for col in int_type_cols:
    
    print(f"{col} : {df[df[col] <= 0].shape}")
```

```python
#inspect before dropping invalid brokered_by
df.loc[df["brokered_by"] <= 0, ["brokered_by", "price", "status", "state"]].head(10)
```

```python
df.loc[df["brokered_by"] == 0, "brokered_by"] = pd.NA
```

```python
#inspect before dropping invalid prices
df.loc[df["price"] <= 0, ["price", "status", "state"]].head(10)
```

```python
df.loc[df["price"] == 0, "price"] = pd.NA
```

```python
df = df.dropna(subset=["price"])
```

**Listings with price equal to zero were reclassified as missing prices. These primarily correspond to ‘ready_to_build’ or incomplete listings and were excluded only for price-based analyses.**

```python
df[df["acre_lot"] <= 0].shape
```

```python
df.loc[df["acre_lot"] == 0, "acre_lot"] = pd.NA
```

```python
df = df.dropna(subset=["house_size"])
```

```python
# Check null ratio
df.isna().mean().sort_values(ascending=False)
df.isna().sum()
```

### Adding new useful columns and information

```python
# Price per Sqft
df["price_per_sqft"] = np.where(
    (df["house_size"].notna()) & (df["house_size"] > 0),
    df["price"] / df["house_size"],
    np.nan
)
```

```python
#Price trends over time by state
df["prev_sold_year"] = df["prev_sold_date"].dt.year.astype('Int64')
```
